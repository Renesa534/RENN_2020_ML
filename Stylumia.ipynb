```
import pandas as pd
```

```
train= pd.read_csv('/content/train.tsv', delimiter= '\t')
train.head()
```

```
train
```

```
train.isnull().sum()
```

```
from sklearn.preprocessing import LabelEncoder
```

```
le= LabelEncoder()
```

```
train['url']= le.fit_transform(train['url'].values)
train['boilerplate']= le.fit_transform(train['boilerplate'].values)
train['alchemy_category']= le.fit_transform(train['alchemy_category'].values)
train['alchemy_category_score']= le.fit_transform(train['alchemy_category_score'].values)
train['is_news']= le.fit_transform(train['is_news'].values)
train['news_front_page']= le.fit_transform(train['news_front_page'].values)
```

```
x= train.iloc[:,0:25]
y= train['label']
```

```
x
```

```
y
```

```
from sklearn.preprocessing import StandardScaler
sc= StandardScaler()
x_scaled= sc.fit_transform(x)
```

```
x_scaled
```

```
from sklearn.model_selection import train_test_split 
```

```
x_train,x_val,y_train,y_val = train_test_split(x_scaled,y,test_size=0.2)
```

DECISION TREE

```
from sklearn.tree import DecisionTreeClassifier
```

```
Dt= DecisionTreeClassifier()
Dt.fit(x_train,y_train)
```

```
print('Training accuracy:', Dt.score(x_train,y_train))
```

```
Dt.score(x_val,y_val)
```

```
from sklearn.metrics import confusion_matrix, classification_report
```

```
import numpy as np
p= Dt.predict(x_val)

print(confusion_matrix(y_val,p))
```

```
p
```

```
print(classification_report(y_val,p))
```

RANDOM FOREST

```
from sklearn.ensemble import RandomForestClassifier
```

```
model= RandomForestClassifier(n_estimators= 50, random_state= 42)
model.fit(x_train, y_train)
```

```
print('Training accuracy:', model.score(x_train,y_train))
```

```
model.score(x_val,y_val)
```

```
p1= model.predict(x_val)
print(confusion_matrix(y_val,p1))
```

```
print(classification_report(y_val,p1))
```

SVM

```
from sklearn.svm import SVC
model_svm= SVC()
model_svm.fit(x_train, y_train)
```

```

print('Training accuracy:', model_svm.score(x_train,y_train))
```

```
model_svm.score(x_val,y_val)
```

```
p3= model_svm.predict(x_val)
print(confusion_matrix(y_val,p3))
```

```
print(classification_report(y_val,p3))
```

KNN

```
from sklearn.neighbors import KNeighborsClassifier
model_knn= KNeighborsClassifier()
model_knn.fit(x_train, y_train)
```

```
print('Training accuracy:', model_knn.score(x_train,y_train))
```

```
model_knn.score(x_val,y_val)
```

```
p2= model_knn.predict(x_val)
print(confusion_matrix(y_val,p2))
```

```
print(classification_report(y_val,p2))
```
